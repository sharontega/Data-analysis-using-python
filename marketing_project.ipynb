```python
import pandas as pd
import numpy as py
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('market.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.columns=df.columns.str.replace(' ','')
```

```python
df.info()
```

```python
df['Income']=df.Income.str.replace('$','')
df['Income']=df.Income.str.replace(',','')
```

```python
df['Income']=df.Income.fillna(df.Income.median())
```

```python
df['Income']=df.Income.astype('float')
df['Income']=df.Income.astype('int')
df['Dt_Customer']=pd.to_datetime(df.Dt_Customer)
```

```python
num=df[['Year_Birth','Income','Dt_Customer', 'Recency', 'MntWines','MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts','MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases','NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth',]]
cat=df[['Education', 'Marital_Status','Kidhome', 'Teenhome','AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1',
       'AcceptedCmp2', 'Response', 'Complain', 'Country']]
```

```python
for i in num:
    sns.histplot(num[i])
    plt.title(i)
    plt.show()
```

```python
for i in cat:
    sns.countplot(cat[i])
    plt.title(i)
    plt.show()
```

```python
df.isnull().sum()
```

```python
a=df[df.Income>200000].index
df.drop(a,inplace=True)
```

```python
df['Age']=2022-df.Year_Birth
df['children']=df.Kidhome+df.Teenhome
df['year']=pd.DatetimeIndex(df['Dt_Customer']).year
df['Dt_Customer_year']=2022-(df.Dt_Customer).dt.year
c=[col for col in df.columns if 'Mnt' in col]
df['Amount']=df[c].sum(axis=1)
d=[col for col in df.columns if 'Accepted' in col]+['Response']
df['Total_accepted']=df[d].sum(axis=1)
e=[col for col in df.columns if 'Purchases' in col]
df['Total_purchase']=df[e].sum(axis=1)
df['Online_purchase']=df.NumWebPurchases+df.NumCatalogPurchases
```
