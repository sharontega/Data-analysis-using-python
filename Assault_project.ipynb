```python
df['occurrencedate']=pd.to_datetime(df.occurrencedate)
df['reporteddate']=pd.to_datetime(df.reporteddate)
```

```python
df.isnull().sum()
```

```python
df['occurrenceyear']=df['occurrenceyear'].fillna(df['occurrencedate'].dt.year)
df['occurrencemonth']=df['occurrencemonth'].fillna(df['occurrencedate'].dt.month)
df['occurrenceday']=df['occurrenceday'].fillna(df['occurrencedate'].dt.day)
```

```python
df['occurrencetime']=df['occurrencedate'].dt.time
df['occurrencedate']=df['occurrencedate'].dt.date
df['reportedtime']=df['reporteddate'].dt.time
df['reporteddate']=df['reporteddate'].dt.date
```

```python
df['occurrenceyear']=df.occurrenceyear.astype('int')
```

```python
df['location_type']=df.location_type.apply(lambda x: x.split('(')[0].split(')')[0])
```

```python
df=df.drop(['X','Y','OBJECTID','Longitude','Latitude','ObjectId2'],axis='columns')              
```
